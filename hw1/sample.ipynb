```python
import numpy as np
from nonlinear import NLSModel
import matplotlib.pyplot as plt
from data import get_sample_data
```

```python
# get the data
train, test, holdout = get_sample_data(100)
```

```python
plt.scatter(train[:, 0], train[:, 1])
plt.show()
```

```python
# get a model up and running
model = NLSModel(0.1, 0.0001, 1, 2)

# fit
train_input = train[:, 0]
train_answers = train[:, 1]
model.fit(train[:, 0], train[:, 1])
```

```python
fig = plt.figure()
scores = model.get_scores()
weights = model.get_weights()
fig.add_subplot(121).plot(scores)
fig.add_subplot(122).plot(weights)
plt.show()
```

```python
values = model.predict(weights[-1], train_input)

plt.scatter(train_input, train_answers)
plt.plot(train_input, values)
```
