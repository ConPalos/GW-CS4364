```python
import numpy as np
from nonlinear import NLSModel
import matplotlib.pyplot as plt
from data import get_sample_data, get_sample_data_3d
```

```python
# get the data
train, test, holdout = get_sample_data(100)
```

```python
plt.scatter(train[:, 0], train[:, 1])
plt.show()
```

```python
# get a model up and running
model = NLSModel(0.0001, 0.000000001, 0, 2)

# fit
train_input = train[:, 0]
train_answers = train[:, 1]
model.fit(train[:, 0], train[:, 1])
```

```python
fig = plt.figure()
scores = model.get_scores()
weights = model.get_weights()
fig.add_subplot(121).plot(scores)
fig.add_subplot(122).plot(weights)
plt.show()
```

```python
values = model.predict(weights[-1], train_input)
plt.scatter(train_input, train_answers)
plt.scatter(train_input, values)
plt.show()
```

```python
test_input = test[:, 0]
test_answers = test[:, 1]

score = model.test(test_input, test_answers)
pred = model.predict(weights[-1], test_input)

print(score)
fig = plt.figure()
fig.add_subplot(121).scatter(test_input, pred, c=(test_answers - pred))
ax = fig.add_subplot(122)
ax.scatter(test_input, test_answers)
ax.scatter(test_input, pred)
plt.show()
```

```python
train, test, holdout = get_sample_data_3d(1000)
plt.scatter(train[:, 0], train[:, 1], c=train[:, 2])
plt.show()
```

```python
model = NLSModel(0.00001, 0.000000001, 0, 2)

# fit
train_input = train[:, :2]
train_answers = train[:, 2]
model.fit(train_input, train_answers)
```

```python
fig = plt.figure()
scores = model.get_scores()
weights = model.get_weights()
fig.add_subplot(221).plot(scores)
fig.add_subplot(222).plot(weights)
values = model.predict(weights[-1], train_input)
fig.add_subplot(223).scatter(train_input[:, 0], train_input[:, 1], c=train_answers)
fig.add_subplot(224).scatter(train_input[:, 0], train_input[:, 1], c=values)
plt.show()
```
