```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# start with train sizes
train = np.arange(1, 6)*100
```

```python
# ok manually set train scores
train_scores = np.array([0.6, 0.5, 0.45, 0.42, 0.4])
test_scores = np.array([0.2, 0.22, 0.23, 0.235, 0.24])
```

```python
# combine into an array
data = pd.DataFrame(columns=['x', 'train', 'test'])
data['x'] = train
data['train'] = train_scores
data['test'] = test_scores
data = data.set_index('x', drop=True)
```

```python
plt.plot(data)
plt.xlabel('Training instances')
plt.ylabel('Score')
plt.show()
```
